```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mainFunction():
    student_scores = [11,14,4,18,3,8,8,5,14,7,2,4,7,9,4,6,10,2,10,16,5,0,19,17,20,13,1,5,19,16,15,2,19,8,10,8,12,15,13,6,2,7,4,19,17,6,7,5,15,5,13,2,15,4,15,1,10,18,1,7,9,6,13,15,9,20,7,15,20,12,3,0,15,17,3,0,18,20,19,2,11,11,19,17,16,18,15,0,9,17,6,12,3,4,8,17,16,17,4,1]
    total_students = len(student_scores)
    return student_scores, total_students

student_scores, total_students = mainFunction()


def calculateMean(student_scores, total_students):
  total_sum = 0
  for score in student_scores:
    total_sum = total_sum + score # finds the total sum os all scores
  mean = total_sum / total_students
  return mean

mean = calculateMean(student_scores,total_students)
print("The Mean is:", round(mean, 2))

def calculateMedian(student_scores, total_students):
    sorted_scores = student_scores[:]# makes a copy of student_scores list

    for i in range(total_students):
      for g in range(total_students - i - 1):# loop sthrought not sorted list
        if sorted_scores[g]> sorted_scores[g + 1]:
          sorted_scores[g], sorted_scores[g + 1] = sorted_scores[g + 1], sorted_scores[g]#swap them in list if it's biger

    middle = total_students // 2
    if total_students %2 == 0:# if even num of students
        median = (sorted_scores[middle - 1] + sorted_scores[middle]) / 2
    else:# if odd num of students
          median = sorted_scores[middle]
    return median

median = calculateMedian(student_scores, total_students)
print("The Median is:", round(median, 2))

def calculateMode(student_scores):
  unique_scores = []
  for score in student_scores:
      if score not in unique_scores:
        unique_scores.append(score)

  # Count frequency of each value
  frequency = []
  for score in unique_scores:
    count = 0
    for student_score in student_scores:
      if student_score == score:
        count += 1
    frequency.append(count)
  # Find the highest frequency
  max_frequency = 0 # starts from the first frequency as max
  for count in frequency:
    if count > max_frequency:
      max_frequency = count
  # Find the modes
  mode = []
  for i in range(len(unique_scores)):
    if frequency[i]==max_frequency:
      mode.append(unique_scores[i])
  return mode
mode = calculateMode(student_scores)
print("The Mode is:", mode)

import math

def standardDev(student_scores, mean, total_students):
    sum_sqr_diff = 0
    for i in student_scores:
        sqr_diff = (i - mean) ** 2 # calculates the squared difference between each score and the mean
        sum_sqr_diff += sqr_diff # finds the sum of all squared differences
    deviation = math.sqrt(sum_sqr_diff / total_students)
    return deviation

deviation = standardDev(student_scores, mean, total_students)
print("The Standard Deviation = ", round(deviation, 2))
```

```python
#Visualisation
plt.hist(student_scores, bins=21, color='lightblue', edgecolor='grey')

plt.axvline(mean, linewidth=1, label="Mean", color='red')
plt.axvline(median, linestyle='dashed', linewidth=1, label="Median", color='green')
for i in mode:
    plt.axvline(i, linestyle='dashed', linewidth=1, label="Mode", color='purple')

plt.axvline(deviation,linestyle='dashed', linewidth=1, label="Sd Deviation", color='blue' )

plt.title("Test Scores")
plt.ylabel("Number of Students")
plt.xlabel("Student Scores")


plt.legend()#little table
plt.grid(True)
plt.show()
```
